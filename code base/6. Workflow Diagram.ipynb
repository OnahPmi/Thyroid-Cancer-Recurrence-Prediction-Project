```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment='Research Workflow')

# Set graph attributes for a wider and shorter layout
dot.attr(size='8.27,5.85', rankdir='TB', splines='ortho')

color = "lightblue"
edgecolor = "red"  # darkturquoise
fontsize='25'

# Add nodes for each step in the workflow with increased font size and color fill
dot.node('A', 'Data Collection and\nPreprocessing', shape='cylinder', style='filled', color=color, fontname='Helvetica-Bold', 
         fontsize=fontsize)
with dot.subgraph(name='cluster_A') as cluster:
    cluster.node('C', 'One-hot encoding\nfor categorical features', shape='box', style='filled', color=color, 
         fontname='Helvetica-Bold', fontsize=fontsize)
    cluster.node('D', 'Minâ€“max normalization\nfor numerical features', shape='box', style='filled', color=color, 
         fontname='Helvetica-Bold', fontsize=fontsize)
dot.node('E', 'Feature Engineering', shape='cylinder', style='filled', color=color, fontname='Helvetica-Bold', 
         fontsize=fontsize)
with dot.subgraph(name='cluster_B') as cluster2:
    cluster2.node('F', 'Linear Dimensionality Reduction\n(PCA, t-ICA, t-SVD, NMF)', shape='box', style='filled', 
             color=color, fontname='Helvetica-Bold', fontsize=fontsize)
    cluster2.node('G', 'Manifold Learning\n(t-SNE, UMAP, LLE, Isomap)', shape='box', style='filled', color=color, 
             fontname='Helvetica-Bold', fontsize=fontsize)
dot.node('H', 'K-means Clustering and Evaluation (10-Fold CV)\nARI, V-measure, Silhouette Coeficient', shape='box', 
         style='filled', color=color, fontname='Helvetica-Bold', fontsize=fontsize)
dot.node('J', 'Classification Pipelines Building\n SVM, LR, FNN, GB, RF, KNN', shape='box', style='filled', color=color, 
         fontname='Helvetica-Bold', fontsize=fontsize)
dot.node('K', 'Classification Pipelines Evaluation on Test Set and in Stratified 10-Fold CV\nBalanced accuracy, AUC, F1 score, etc.', 
         shape='box', style='filled', color=color, fontname='Helvetica-Bold', fontsize=fontsize)
dot.node('L', 'Classification Pipelines Comparison (Wilcoxon Signed-Rank Test ) and\nSelection of Best Performing Classification Pipeline', 
         shape='box', style='filled', color=color, fontname='Helvetica-Bold', fontsize=fontsize)
dot.node('M', 'Stratified Subgroup Evaluation of \nthe Best Performing Classification Pipeline', 
         shape='box', style='filled', color=color, fontname='Helvetica-Bold', fontsize=fontsize)
dot.node('N', 'Explainability Analysis of the Best Performing Classification Pipeline\nusing the SHAP Values', 
         shape='cylinder', style='filled', color=color, fontname='Helvetica-Bold', fontsize=fontsize)


edgeLen = "1.7"
# Add edges to define the flow with increased arrow size
dot.edge('A', 'C', arrowsize=edgeLen, color=edgecolor)
dot.edge('A', 'D', arrowsize=edgeLen, color=edgecolor)
dot.edge('C', 'E', arrowsize=edgeLen, color=edgecolor)
dot.edge('D', 'E', arrowsize=edgeLen, color=edgecolor)
dot.edge('E', 'F', arrowsize=edgeLen, color=edgecolor)
dot.edge('E', 'G', arrowsize=edgeLen, color=edgecolor)
dot.edge('F', 'H', arrowsize=edgeLen, color=edgecolor)
dot.edge('G', 'H', arrowsize=edgeLen, color=edgecolor)
dot.edge('H', 'J', arrowsize=edgeLen, color=edgecolor)
dot.edge('J', 'K', arrowsize=edgeLen, color=edgecolor)
dot.edge('K', 'L', arrowsize=edgeLen, color=edgecolor)
dot.edge('L', 'M', arrowsize=edgeLen, color=edgecolor)
dot.edge('M', 'N', arrowsize=edgeLen, color=edgecolor)

# Render the graph to a file
dot.render('../results/research_workflow_wide 1', format='png', cleanup=True)

# Display the graph
dot
```
